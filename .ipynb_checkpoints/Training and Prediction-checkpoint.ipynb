```python
import pandas as pd
```

```python
legitimate = pd.read_csv("legitimate-urls.csv")
```

```python
phishing = pd.read_csv("phishing-urls.csv")
```

```python
legitimate.head(10)
```

```python
phishing.head(10)
```

```python
merged = legitimate.append(phishing)
```

```python
merged.head(10)
```

```python
merged.columns
```

```python
merged = merged.drop(merged.columns[[0,3,5]],axis=1)
```

```python
merged = merged.sample(frac = 1)
```

```python
urls_without_labels = merged.drop('label',axis=1)
urls_without_labels.columns
labels = merged['label']
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_labels, test_labels = train_test_split(urls_without_labels, labels, test_size=0.20, random_state=100)
```

```python
print(len(train_data),len(test_data),len(train_labels),len(test_labels))
```

```python
#Decision tree

from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier()
```

```python
tree.fit(train_data,train_labels)
```

```python
pred = tree.predict(test_data)
```

```python
print(pred)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(test_labels,pred)
print(cm)
accuracy_score(test_labels,pred)
```

```python
#saving the model

from joblib import dump, load
dump(tree, 'decisiontree.joblib')
tree = load('decisiontree.joblib')
```

```python
# using pickle

import pickle 
pickle_out = open("classifier.pkl", mode = "wb") 
pickle.dump(tree, pickle_out) 
pickle_out.close()
```
